```python
import pandas as pd
df_in = pd.read_csv("modified clean_data.csv")
df_us = pd.read_csv("modified_data.csv")
```

```python
df_in.columns = df_in.columns.str.lower().str.replace(' ', '_')
df_us.columns = df_us.columns.str.lower().str.replace(' ', '_')
```

```python
df_combined = pd.concat([df_in, df_us], ignore_index=True)
df_combined
```

```python
from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df_combined['title_sentiment'] = df_combined['title'].apply(get_sentiment)
df_combined['tags_sentiment'] = df_combined['tags'].apply(get_sentiment)
```
